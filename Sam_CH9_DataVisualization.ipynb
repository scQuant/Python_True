### Annotations and Drawings on Subplot

```python
from datetime import datetime

fig = plt.figure(figsize = (12,8))
ax = fig.add_subplot(1,1,1)

data = pd.read_csv('examples/spx.csv', index_col = 0, parse_dates = True)  # This will treat index column using the 1st column and also PARSE dates! 
spx = data['SPX']
spx.plot(ax = ax, style = 'k--') # Setting ax = ax will bind the dataframe plot to the created subplot
ax.legend(loc = 'best')

crisis_data = [
(datetime(2007, 10, 11), 'Peak of bull market'),
(datetime(2008, 3, 12), 'Bear Stearns Fails'),
(datetime(2008, 9, 15), 'Lehman Bankruptcy')
]

for date, label in crisis_data:
    ax.annotate(label, xy = (date, spx.asof(date) + 75), xytext = (date, spx.asof(date) + 150), arrowprops = dict(facecolor = 'black'))

ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])
ax.set_title('Example')

rect = plt.Rectangle((0,200), 100, 100)
ax.add_patch(rect) # Add patch to add shapes to the graph, although likly needs to be numerical axis...

plt.savefig('Example_graph') # Used to save the graph
```

# Plotting with Matplotlib and Seaborn

```python
import seaborn as sns

tips = pd.read_csv('examples/tips.csv')
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])

tips.head()

#Seaborn bar plots
sns.barplot(x = 'tip_pct', y = 'day', data = tips)   # Use column names as x and y, and have a DATA argument
sns.barplot(x = 'tip_pct', y = 'day', hue = 'smoker', data = tips)  # HUE gives an additional categorical filter

#Seaborn hist and density plots
#sns.distplot(tips['total_bill'], bins = 50, color = 'b')

#Seaborn Factor Plot
sns.factorplot(x='day', y='tip_pct', hue='time', col='smoker', kind='bar', data=tips) # Breaking down data by factors! 
```

```python
sns.set(style = 'whitegrid') # Can set the style of seaborn

#Scatter/Reg plot
sns.regplot('total_bill', 'tip_pct', data = tips)

#Pair Plot (scatter plot for all pairs!)
sns.pairplot(tips[['total_bill','tip', 'size']]) 
```
