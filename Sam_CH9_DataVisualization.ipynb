### Figures and Subplots

```python
fig = plt.figure(figsize = (15,10)) # plots in plt resides within a figure
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.hist(np.random.randn(500), bins = 10, color = 'r', alpha = 0.2)
```

```python
fag, axe = plt.subplots(2,2, figsize = (15,10), sharex = True) # Create the figure and subplots directly!!!
axe[0,1].hist(np.random.randn(500), bins = 10, color = 'r', alpha = 0.2)
plt.subplots_adjust(wspace = 0.1, hspace = 0.1) # Making sure no space in between 
plt.legend(loc = 'best')
axe[0,1].set_yticks([0,25,50,75,100])
axe[0,1].set_title('Tadaxi')
                   
props = {'title': 'Tadaxii', 'yticks': [0,25,50,75,100,125]} # Creating a dictionary of property to be set together!
axe[0,0].set(**props)
```
